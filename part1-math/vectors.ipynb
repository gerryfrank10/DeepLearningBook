```python
# Transpose
import math
import numpy as np
import matplotlib.pyplot as plt
v = np.array([2, 5, 1, 7])
v_T = v.T
v_T, v
```

```python
# How about row vector?
v_row = v.reshape(1, -1)
v_row, v_row.T
```

```python
# Vector * Number
v = np.array([2, 3]); v0 = 3 * v # alpha=3
plt.plot((0, v0[0]), (0, v0[1]), 'r',label=r'$3 \cdot v$')
plt.plot((0, v[0]), (0, v[1]), 'b',label=r'$v$')
plt.legend()
```

```python
# Vector Multiplication
v = np.array([1j, 3])
w = np.array([3j, 2])
print(v @ w)
print(np.dot(v, w))
```

```python
# Module of vector
# ||v|| = sqrt(v @ v)
v = np.array([2, 3])
print(np.sqrt(v @ v))
print(np.linalg.norm(v))
plt.axline((0, 0), (2, 3))
```

```python
# Vector Unitary
v = np.array([2, np.pi **(np.exp(2)), 0])
u = v / np.linalg.norm(v)
print(u), print(np.linalg.norm(u))
```

```python
import math , numpy as np, matplotlib as mpl
from mpl_toolkits.mplot3d import axes3d
w = np.array([3, 4,6])
print(np.sqrt(w @ w)); print(np.linalg.norm(w))
ax = plt.axes(projection='3d')
ax.plot([0, 3], [0, 4], [0, 6], 'b--')
ax.plot([0, 3], [0, 4], [0, 0], 'r--')
ax.text(0, 3, 3, r'$\sqrt{3^2+4^2+6^2}$', color='b')
ax.text(1, 0, 0, r'$\sqrt{(3-0)^2+(4-0)^2}$', color='r')
```

```python
# Angle between two vectors
a = np.array([2j, 0.5, -1]); b = np.array([0, 3, 3/2])
cos_theta = (a @ b) / (np.linalg.norm(a) * np.linalg.norm(b))
theta = np.arccos(cos_theta)
print(f'Cosine of the angle: {cos_theta}')
print(f'Angle in radians: {theta}')
```

```python
ax = plt.axes(projection='3d')
ax.plot([0, 2j], [0, 0.5], [0, -1], 'b--', label='Vector a')
ax.plot([0, 0], [0, 3], [0, 3/2], 'r--', label='Vector b')
ax.text(0, 0.5, -1, f'$\\theta={np.round(theta,2 )}$', color='b')
ax.legend()
```
